> **Net Migration** refers to the difference between the number of people entering a region (immigrants) and the number of people leaving the region (emigrants) over a specific period of time. It reflects the net effect of migration on the population of the region.

> **Mathematical Formula for Migration**

$$
\text{Net Migration} = \text{Number of Immigrants} - \text{Number of Emigrants}
$$
Where,
- `Immigrants` are the people entering a region (moving into the area)
- `Emigrants` are the people going out of a region (moving out of the area)  

> **Mathematical Formula for Net Migration Rate**

$$\text{Net Migration Rate(per 1000 people)} = \frac{\text{Number of Immigrants} - \text{Number of Emigrants}}{\text{Total Population}} \times 1000$$

```python
#### We want to change the column name from Years to date
data_migration.rename(columns={"Years":"date"}, inplace=True)
data_migration.info()
```

```python
##### We want to add country column 
# Create a Series with 64 rows of "India" and 64 rows of "Pakistan"
country_series = pd.Series(["India"] * 64 + ["Pakistan"] * 64, name="country")
years = list(range(1960, 2024))
# Repeat the years list twice (once for India, once for Pakistan)
date_series = pd.Series(years * 2, name="date")
netMigration_series = pd.Series(
    pd.concat([data_migration['India_netMigration'], data_migration['Pakistan_netMigration']], ignore_index=True),
    name="netMigration"
)

# Combine the Series into a DataFrame
migration_data = pd.DataFrame({
    'date': date_series,
    'country': country_series,
    'netMigration': netMigration_series
})


#### data_population contains the population data and data_migration contains the Net migration data. 
#### Lets merge them based on the date and country columns ###################################
pop_migration_ind_pk = pd.merge(data_population, migration_data, on=['date', 'country'], how='inner')
pop_migration_ind_pk.info()
```

```python


# Convert population to millions
pop_migration_ind_pk['population_millions'] = pop_migration_ind_pk['population'] / 1_000_000

# Scatter plot with regression line for each country
plt.figure(figsize=(12, 6))

sns.regplot(
    data=pop_migration_ind_pk[pop_migration_ind_pk['country'] == "India"],
    x="population_millions",
    y="netMigration",
    label="India",
    scatter_kws={'alpha': 0.6},  # Adjust transparency of points
    line_kws={'color': 'blue'}  # Line color
)

sns.regplot(
    data=pop_migration_ind_pk[pop_migration_ind_pk['country'] == "Pakistan"],
    x="population_millions",
    y="netMigration",
    label="Pakistan",
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'green'}
)

plt.title("Correlation Between Population (in Millions) and Net Migration")
plt.xlabel("Population (in Millions)")
plt.ylabel("Net Migration")
plt.legend(title="Country")
plt.grid()
plt.tight_layout()
plt.show()
```
